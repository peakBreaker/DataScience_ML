```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

### Simple functional plotting

The simplest way to get started with plotting is to use the functinal plotting way

```python
## FUNCTIONAL
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title("My plot")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### Object oriented paradigm

Matplotlib supports an object oriented way to specify the visualization, which is much better to work with

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X LABEL')
axes.set_title('My title')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.2,0.4,0.4,0.3])

# and we can plot in the axes
ax1.plot(x,y)
ax1.set_title("BIG PLOT")
ax2.plot(y,x)
ax2.set_title("SMALL PLOT")
```

#### OO Subplotting

```python
fig, axes = plt.subplots(nrows=3,ncols=3)
fig.tight_layout() ## Use to avoid overlapping labels
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

## So axes is just a list of axes- So we can iterate through axes
for curr_ax in axes:
    curr_ax.plot(x,y)
```

## Figure size and DPI

With matplotlib we can alter and specify attributes of our plots - we'll take a look at a couple of them here:

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(x,y)

plt.tight_layout()
## Now we may save the figure with DPI specified
fig.savefig('myplot.png', dpi=200)
```

### Legends

To label the plots we may use legends:

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = "X Squared")
ax.plot(x,x**3, label = "X Cubed")

ax.legend(loc=10) ## Location codes - 0 is default as "smart"
                  ## Can also be a location (x,y) tuple
```

### Plot Appearance

Lets make our plots beautiful!

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,
        color='#000000', # Can take hex values
        lw=3, # Linewidth
        ls='steps', # Linestyle
        marker='o', # Add markers
        markersize=20,
        markerfacecolor='yellow', # Color of the markers
        markeredgewidth=3,
        alpha=0.5)
## Basically just look up the documentation and go nuts
```

```python
## Setting lims
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=2, ls='--')
ax.set_xlim([0,10])
ax.set_ylim([0,2])
```
