```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Distribution plots

Lets take a look at some simple distribution plots.  Lets start with a histogram:

```python
## Level 1 - Simple histogram
sns.distplot(tips['total_bill'], kde=False, bins=40)
```

```python
## Level 2 - Joint plots
# Two numerical columns together will create a scatter plot, where we 
# can see the correlations between the values
# Substitute kind with 'reg', 'hex' or 'kde'
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
## Level 3 - Pairplots
# Pairplots will take a dataset and plot all numerical values against 
# each other. We may also color hue using a non-numerical value
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

## KDE
KDE or Kernel Density Estimation is a way of showing the distribution of data.  It is made using the rugplots (below), where each point creates a normal dist curve, and the KDE plot is generated by accumulating these together

https://en.wikipedia.org/wiki/Kernel_density_estimation

```python
sns.rugplot(tips['total_bill'])
```
