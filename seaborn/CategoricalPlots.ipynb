## Categorical plots
In this lecture, we take a closer look at plots we can use to compare categories:
- Barplots
- Boxplots
- Violinplots (my fav)
- Strip&swarm plots

Lets start by importing some sample data:

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
## Boxplots are good for data presentation, but can hide some underlying data
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
## Violinplots can be very powerful for comparing data and exploring data
# But may not be optimal for presenting data as many are not familiar with
# violinplots
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
## Strip plots can also be very useful, but use with care
# Note the use of jitter, split and hue
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True)
```

```python
## Swarmplot- Good for small datasets and data exploration
# We can also stack plots on top of each other
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```
